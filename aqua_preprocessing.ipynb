```python
import pandas as pd

# Assuming df is your DataFrame with a 'Label' column containing categorical labels
df = pd.DataFrame({'Label': ['A', 'B', 'C', 'A', 'B', 'C', 'D', 'E', 'D']})

# Use factorize to map categorical labels to numerical values
df['Label'], label_mapping = pd.factorize(df['Label'])

# Now, df['Label'] contains numerical values corresponding to the categorical labels
print(df)

# Create a mapping dictionary
mapping_dict = dict(zip(label_mapping, range(len(label_mapping))))

# Print the mapping dictionary
print("Mapping Dictionary:")
print(mapping_dict)
```
